```python
pq_page = pq(page.text)
if pq_page('table').eq(0):
    data = [[pq(child).text() for child in pq(tr).children()] for tr in pq_page('table').eq(0).find("tr") if(len(pq(tr).text()))]
df = pd.DataFrame(data=data[1:-1], columns=data[0])
```
